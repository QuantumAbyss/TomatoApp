```python
from fastai.vision.all import *
from fastai.vision.widgets import ImageClassifierCleaner
from azure.cognitiveservices.search.imagesearch import ImageSearchClient as api
from msrest.authentication import CognitiveServicesCredentials as auth
```

```python
def search_images_bing(key, term, min_sz=128):
    client = api('https://api.cognitive.microsoft.com', auth(key))
    return L(client.images.search(query=term, count=150, min_height=min_sz, min_width=min_sz).value)
```

# Get Tomato Data

```python
tomato_types = 'wild red cherry','pink brandywine','san marzano','cherokee purple','green zebra','amish plum','yellow pear'
path = Path('tomatoes')

if not path.exists():
    path.mkdir()
    for o in tomato_types:
        dest = (path/o)
        dest.mkdir(exist_ok=True)
        results = search_images_bing(key, f'{o} tomato')
        download_images(dest, urls=results.attrgot('content_url'))
```

```python
fns = get_image_files(path)
fns
```

```python
failed = verify_images(fns)
failed
```

```python
failed.map(Path.unlink);
```

# Build Model

```python
tomatoes = DataBlock(
    blocks=(ImageBlock, CategoryBlock),
    get_items=get_image_files,
    splitter=RandomSplitter(valid_pct=0.3, seed=25),
    get_y=parent_label,
    item_tfms=RandomResizedCrop(128, min_scale=0.2),
    batch_tfms=aug_transforms(mult=2))
#tomatoes = tomatoes.new(item_tfms=Resize(128),
#                       batch_tfms=aug_transforms(mult=2))
```

```python
dls = tomatoes.dataloaders(path)
```

```python
dls.train.show_batch(max_n=8, nrows=2, unique=True)
```

# Training

```python
learn = cnn_learner(dls, resnet34, metrics=error_rate)
learn.fine_tune(7)
```

```python
interp = ClassificationInterpretation.from_learner(learn)
interp.plot_confusion_matrix()
```

```python
interp.plot_top_losses(9, nrows=3)
```

```python
cleaner = ImageClassifierCleaner(learn)
cleaner
```

```python
for idx in cleaner.delete(): cleaner.fns[idx].unlink()
for idx,cat in cleaner.change(): shutil.move(str(cleaner.fns[idx]), path/cat)
```
